```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast

import warnings
warnings.filterwarnings('ignore')
```

```python
book_data = pd.read_csv('data/books_data.csv')
book_data
```

```python
book_rating = pd.read_csv('data/Books_rating.csv')
book_rating
```

```python
book_data['authors'] = book_data['authors'].fillna('[]')
book_data['authors'] = book_data['authors'].apply(ast.literal_eval)
merged_data = pd.merge(book_rating, book_data[['Title', 'authors']], on='Title', how='left')
```

```python
merged_data
```

```python
merged_data['review'] = merged_data['summary'].fillna('') + ' ' + merged_data['text'].fillna('')
```

### Get the sentiment analysis

```python
from textblob import TextBlob

def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

merged_data['sentiment'] = merged_data['review'].apply(get_sentiment)
```

```python
merged_data = merged_data.explode('authors')

author_sentiment = merged_data.groupby('authors')['sentiment'].mean().reset_index()
author_sentiment.columns = ['Author', 'Average Sentiment']

print(author_sentiment)
```

### Visualizing the top 20 and the bottom 20 authors by average sentiment review

```python
sns.set_context("poster")
plt.figure(figsize=(20, 15))

sns.barplot(x='Average Sentiment', y='Author', data=author_sentiment.sort_values(by='Average Sentiment', ascending=False).head(20), palette='coolwarm')
plt.title("Top 20 Authors by Average Sentiment of Reviews")
plt.xlabel("Average Sentiment")
plt.ylabel("Author")
plt.show()
```

```python
sns.set_context("poster")
plt.figure(figsize=(20, 15))

sns.barplot(x='Average Sentiment', y='Author', data=author_sentiment.sort_values(by='Average Sentiment', ascending=True).head(20), palette='coolwarm')
plt.title("Bottom 20 Authors by Average Sentiment of Reviews")
plt.xlabel("Average Sentiment")
plt.ylabel("Author")
plt.show()
```

### Function go get the sentiment of an author

```python
def get_author_sentiment(author_name):
    author_reviews = merged_data[merged_data['authors'] == author_name]
    if author_reviews.empty:
        return f"No reviews found for author: {author_name}"
    average_sentiment = author_reviews['sentiment'].mean()
    return f"The average sentiment for author {author_name} is {average_sentiment:.2f}"

# Example usage
print(get_author_sentiment("J.K. Rowling"))
print(get_author_sentiment("George Orwell"))
print(get_author_sentiment("William Shakespeare"))
print(get_author_sentiment("Charles Chase Lord"))
print(get_author_sentiment("Willard Motley"))
```

### Investigating the number of reviews per author

```python
def get_author_sentiment_and_number_of_reviews(author_name):
    author_reviews = merged_data[merged_data['authors'] == author_name]
    num_reviews = author_reviews.shape[0]
    if num_reviews == 0:
        return f"No reviews found for author: {author_name}"
    average_sentiment = author_reviews['sentiment'].mean()
    return f"The average sentiment for author {author_name} is {average_sentiment:.2f}, based on {num_reviews} reviews."

# Example usage
print(get_author_sentiment_and_number_of_reviews("J.K. Rowling"))
print(get_author_sentiment_and_number_of_reviews("George Orwell"))
print(get_author_sentiment_and_number_of_reviews("William Shakespeare"))
print(get_author_sentiment_and_number_of_reviews("Charles Chase Lord"))
print(get_author_sentiment_and_number_of_reviews("Willard Motley"))
```

```python
author_sentiment_with_number_reviews = merged_data.groupby('authors').agg(
    Average_Sentiment=('sentiment', 'mean'),
    Number_of_Reviews=('sentiment', 'size')
).reset_index()

print(author_sentiment_with_number_reviews)
```

```python
sns.set_context("poster")
plt.figure(figsize=(20, 15))

top_20_authors = author_sentiment_with_number_reviews.sort_values(by='Average_Sentiment', ascending=False).head(20)
sns.barplot(x='Average_Sentiment', y='authors', data=top_20_authors, palette='coolwarm')
for i, (avg_sentiment, num_reviews) in enumerate(zip(top_20_authors['Average_Sentiment'], top_20_authors['Number_of_Reviews'])):
    plt.text(avg_sentiment, i, f'  {num_reviews} reviews', va='center')
plt.title("Top 20 Authors by Average Sentiment of Reviews")
plt.xlabel("Average Sentiment")
plt.ylabel("Author")
plt.show()
```

```python
sns.set_context("poster")
plt.figure(figsize=(20, 15))

bottom_20_authors = author_sentiment_with_number_reviews.sort_values(by='Average_Sentiment', ascending=True).head(20)
sns.barplot(x='Average_Sentiment', y='authors', data=bottom_20_authors, palette='coolwarm')
for i, (avg_sentiment, num_reviews) in enumerate(zip(bottom_20_authors['Average_Sentiment'], bottom_20_authors['Number_of_Reviews'])):
    plt.text(avg_sentiment, i, f'  {num_reviews} reviews', va='center')
plt.title("Bottom 20 Authors by Average Sentiment of Reviews")
plt.xlabel("Average Sentiment")
plt.ylabel("Author")
plt.show()
```

## Calculating the percentage of authors with small number of reviews

```python
author_review_counts = merged_data.groupby('authors').size().reset_index(name='Number_of_Reviews')
authors_with_one_review = author_review_counts[author_review_counts['Number_of_Reviews'] == 1].shape[0]
total_authors = author_review_counts.shape[0]
portion_with_one_review = authors_with_one_review / total_authors

print(f"Portion of authors with only one review: {portion_with_one_review:.2%}")
```

```python
authors_with_5_or_less_reviews = author_review_counts[author_review_counts['Number_of_Reviews'] <= 5].shape[0]
total_authors_5 = author_review_counts.shape[0]
portion_with_5_or_less_reviews = authors_with_5_or_less_reviews / total_authors_5

print(f"Portion of authors with only one review: {portion_with_5_or_less_reviews:.2%}")
```

```python
authors_with_10_or_less_reviews = author_review_counts[author_review_counts['Number_of_Reviews'] <= 10].shape[0]
total_authors_10 = author_review_counts.shape[0]
portion_with_10_or_less_reviews = authors_with_10_or_less_reviews / total_authors_10

print(f"Portion of authors with only one review: {portion_with_10_or_less_reviews:.2%}")
```

```python
author_review_counts = merged_data.groupby('authors').size().reset_index(name='Number_of_Reviews')

author_review_counts = author_review_counts.sort_values(by='Number_of_Reviews', ascending=False)

author_review_counts['Cumulative_Reviews'] = author_review_counts['Number_of_Reviews'].cumsum()

total_reviews = author_review_counts['Number_of_Reviews'].sum()
author_review_counts['Cumulative_Percentage'] = (author_review_counts['Cumulative_Reviews'] / total_reviews) * 100

plt.figure(figsize=(20, 10))
sns.lineplot(x=range(1, len(author_review_counts) + 1), y='Cumulative_Percentage', data=author_review_counts)
plt.title('Cumulative Percentage of Reviews by Authors')
plt.xlabel('Authors (sorted by number of reviews)')
plt.ylabel('Cumulative Percentage of Reviews')
plt.show()
```
