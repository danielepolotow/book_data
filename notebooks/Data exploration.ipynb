## Books with most ratings

```python
sns.set_context("poster")
plt.figure(figsize = (20,15))
books = book_rating['Title'].value_counts()[:20]
sns.barplot(x=books, y=books.index, palette='deep')
plt.title("Books with most ratings")
plt.xlabel("Number of occurrences")
plt.ylabel("Books")
# plt.tight_layout()
# plt.savefig("top_books.png")
plt.show()
```

## Distribution of Ratings

```python
sns.set_context("poster")
plt.figure(figsize=(20, 15))
sns.histplot(book_rating['score'], bins=5, kde=False, palette='deep', discrete=True)
plt.title("Distribution of Book Scores")
plt.xlabel("Score")
plt.ylabel("Number of Ratings")
# plt.savefig("dist_scores.png")
plt.show()
```

## Authors with most books

```python
book_data['authors'] = book_data['authors'].fillna('[]')
book_data['authors'] = book_data['authors'].apply(ast.literal_eval)
exploded_authors = book_data.explode('authors')
```

```python
author_counts = exploded_authors['authors'].value_counts().reset_index()
author_counts.columns = ['Author', 'Number of Books']

top_authors = author_counts.head(20)

sns.set_context("poster")
plt.figure(figsize = (20,15))
sns.barplot(x='Number of Books', y='Author', data=top_authors, palette='deep')
plt.title("Top 10 Authors with Most Books")
plt.xlabel("Number of Books")
plt.ylabel("Authors")
plt.show()
```

## Top Rated Books

```python
top_rated_books = book_rating[book_rating['score'] == 5]
top_rated_books_count = top_rated_books['Title'].value_counts().head(20).reset_index()
top_rated_books_count.columns = ['Title', 'Number of 5 Scores']
```

```python
sns.set_context("poster")
plt.figure(figsize=(20, 15))
sns.barplot(x='Number of 5 Scores', y='Title', data=top_rated_books_count, palette='deep')
plt.title("Top 20 Books with the Most Number of 5 Scores")
plt.xlabel("Number of 5 Scores")
plt.ylabel("Book Title")
# plt.tight_layout()
# plt.savefig("top_books_scores.png")
plt.show()
```

## Distrbution of books based on genre

```python
book_data = pd.read_csv('original_data/books_data.csv')
book_data['categories'] = book_data['categories'].fillna('[]')
book_data['categories'] = book_data['categories'].apply(ast.literal_eval)
exploded_categories = book_data.explode('categories')
unique_categories = exploded_categories['categories'].unique()
num_unique_categories = len(unique_categories)
print(f'There are {num_unique_categories} unique categories in the dataset.')
```

```python
category_counts = exploded_categories['categories'].value_counts()
categories_once = category_counts[category_counts <= 10]
num_categories_once = len(categories_once)
print(f'There are {num_categories_once} unique categories that appear 10 or less times in the dataset.')
```

```python
category_counts = exploded_categories['categories'].value_counts().reset_index()
category_counts.columns = ['Genre', 'Number of Books']
top_20_categories = category_counts.head(20)
plt.figure(figsize=(20, 15))
plt.pie(
    top_20_categories['Number of Books'],
    labels=top_20_categories['Genre'],
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette("deep", len(top_20_categories)),
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},
    textprops={'fontsize': 12},
    labeldistance=1.1
)
plt.title("Distribution of Top 20 Book Genres", fontsize=20, pad=20)
plt.axis('equal')
# plt.tight_layout()
# plt.savefig("categories.png")
plt.show()
```

```python
sns.set_context("poster")
plt.figure(figsize=(20, 15))
category_counts = exploded_categories['categories'].value_counts().reset_index()
category_counts.columns = ['Genre', 'Number of Books']
top_20_categories = category_counts.head(20)

sns.barplot(
    x='Number of Books', 
    y='Genre', 
    data=top_20_categories, 
    palette='deep'
)

plt.title("Distribution of Top 20 Book Genres", fontsize=20)
plt.xlabel("Number of Books", fontsize=16)
plt.ylabel("Genre", fontsize=16)
# plt.tight_layout()
# plt.savefig("categories_barplot.png")
plt.show()
```

## Wordcloud for the top 20 books with more ratings

```python
from wordcloud import WordCloud

book_rating = pd.read_csv('original_data/Books_rating.csv')
top_20_books = book_rating['Title'].value_counts().head(20).index
top_20_reviews = book_rating[book_rating['Title'].isin(top_20_books)]['text']
all_reviews = ' '.join(top_20_reviews.dropna())
wordcloud = WordCloud(width=1600, height=800, background_color='white').generate(all_reviews)
plt.figure(figsize=(20, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Reviews of Top 20 Books with Most Ratings', fontsize=20)
plt.show()
```

## Top 10 genres with more positive reviews and bottom 10 genres with more negative reviews

```python
book_data = pd.read_csv('original_data/books_data.csv')
book_rating = pd.read_csv('original_data/Books_rating.csv')

book_data['categories'] = book_data['categories'].fillna('[]')
book_data['categories'] = book_data['categories'].apply(ast.literal_eval)
merged_data = pd.merge(book_rating, book_data[['Title', 'categories']], on='Title', how='left')
merged_data = merged_data.explode('categories')

positive_reviews = merged_data[merged_data['score'] == 5]
negative_reviews = merged_data[merged_data['score'] == 1]

positive_category_counts = positive_reviews['categories'].value_counts().reset_index()
positive_category_counts.columns = ['Category', 'Number of Positive Reviews']

negative_category_counts = negative_reviews['categories'].value_counts().reset_index()
negative_category_counts.columns = ['Category', 'Number of Negative Reviews']

top_positive_categories = positive_category_counts.head(10)
top_negative_categories = negative_category_counts.head(10)
```

```python
sns.set_context("poster")
plt.figure(figsize=(20, 15))

plt.subplot(1, 2, 1)
sns.barplot(x='Number of Positive Reviews', y='Category', data=top_positive_categories, palette='coolwarm')
plt.title("Top 10 Categories with Most Positive Reviews")
plt.xlabel("Number of Positive Reviews")
plt.ylabel("Category")

plt.subplot(1, 2, 2)
sns.barplot(x='Number of Negative Reviews', y='Category', data=top_negative_categories, palette='coolwarm')
plt.title("Top 10 Categories with Most Negative Reviews")
plt.xlabel("Number of Negative Reviews")
plt.ylabel("Category")

plt.tight_layout()
plt.show()
```
